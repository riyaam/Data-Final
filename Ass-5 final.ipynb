```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load the dataset
raw_data = pd.read_csv('Survey.csv')
raw_data.head()
```

```python
#Creating Dataframe
raw_data = raw_data[raw_data.self_employed=='Yes']
raw_data["self_employed"] = raw_data["self_employed"].replace(["Yes", "Don't know", "No"], [1, 0.5, 0.1])
raw_data["wellness_program"] = raw_data["wellness_program"].replace(["Yes", "Don't know", "No"], [1, 0.5, 0.1])
raw_data.groupby(['Country','self_employed','wellness_program']).size().reset_index()
```

```python
#Visualization of country and wellness program
raw_data.plot.bar(x='Country',y='wellness_program')
```
